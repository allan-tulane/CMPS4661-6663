```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# read the input image
img = cv2.imread("./res/cat.jpg")
# convert from BGR to RGB so we can plot using matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# disable x & y axis
plt.axis('off')
# show the image
plt.imshow(img)
plt.show()

# get the image shape
rows, cols, dim = img.shape

#angle from degree to radian
angle = np.radians(10)
#transformation matrix for Rotation
M_ro = np.float32([[np.cos(angle), -(np.sin(angle)), 0],
            	[np.sin(angle), np.cos(angle), 0]])


rotated_img0 = cv2.warpAffine(img,M_ro,(int(cols*1.5),int(rows*1.5)))

# disable x & y axis
plt.axis('off')
# show the resulting image
plt.imshow(rotated_img0)

# plt.savefig('rotated_img.png')

plt.show()
```

```python
#transformation matrix for Rotation
M_r = np.float32([[np.cos(angle), -(np.sin(angle)), 0],
            	[np.sin(angle), np.cos(angle), 0],
            	[0, 0, 1]])

# apply a perspective transformation to the image

rotated_img = cv2.warpPerspective(img, M_r, (int(cols*1.5),int(rows*1.5)))
# disable x & y axis
plt.axis('off')
# show the resulting image
plt.imshow(rotated_img)
plt.show()
```

```python
#transformation matrix for Scaling
M_s = np.float32([[1.5, 0  , 0],
            	[0,   1.8, 0],
            	[0,   0,   1]])
# apply a perspective transformation to the image
scaled_img = cv2.warpPerspective(img, M_s,(int(cols*2),int(rows*2)))
# disable x & y axis
plt.axis('off')
# show the resulting image
plt.imshow(scaled_img)
plt.show()
```

```python
# transformation matrix for Shearing
# shearing applied to x-axis
M_h = np.float32([[1, 0.5, 0],
             	[0, 1  , 0],
            	[0, 0  , 1]])
# shearing applied to y-axis
# M = np.float32([[1,   0, 0],
#             	  [0.5, 1, 0],
#             	  [0,   0, 1]])

# apply a perspective transformation to the image                
sheared_img = cv2.warpPerspective(img,M_h,(int(cols*1.5),int(rows*1.5)))
# disable x & y axis
plt.axis('off')
# show the resulting image
plt.imshow(sheared_img)
plt.show()
```

```python
# transformation matrix for translation
M_t = np.float32([[1, 0, 50],
                [0, 1, 50],
                [0, 0, 1]])
# apply a perspective transformation to the image
translated_img = cv2.warpPerspective(img, M_t, (int(cols*1.5),int(rows*1.5)))
# disable x & y axis
plt.axis('off')
# show the resulting image
plt.imshow(translated_img)
plt.show()
```

```python
M_c = np.dot(M_h, M_s)
M_c = np.dot(M_c, M_r)

print(M_c)

combine_img = cv2.warpPerspective(img, M_c, (int(cols*3),int(rows*3)))
# disable x & y axis
plt.axis('off')
# show the resulting image
plt.imshow(combine_img)
plt.show()
```
